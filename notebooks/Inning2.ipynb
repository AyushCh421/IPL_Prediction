```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/ipl.csv")
```

```python
df.head(7)
```

```python
df.drop(columns=['wicket_kind','player_out','fielders','date','match_type','event_name','bat_pos','batter','bowler','runs_batter','balls_faced'],inplace=True)
```

```python
df.shape
```

```python
df.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
df.drop(columns=['valid_ball','runs_extras','runs_not_boundary','extra_type','non_striker','non_striker_pos','review_batter',"team_reviewed",'review_decision',"umpire",'umpires_call','player_of_match','toss_winner',"toss_decision","venue","day",'month','year','season','gender',"team_type",'superover_winner','result_type','method','balls_per_over','overs','event_match_no',"stage","match_number",'new_batter','batter_runs','batter_balls','bowler_wicket','batting_partners','next_batter','striker_out','win_outcome'],inplace=True)
```

```python
df.head()
```

```python
df=df[df['innings']==2]
df.rename(columns={'match_won_by':'winner'},inplace=True)
df.drop(columns=['ball'],inplace=True)
```

```python
df.sample(3)
```

```python
df.team_balls.unique()
```

```python
df.duplicated().sum()
```

```python
df.rename(columns={'team_runs':'current_score'},inplace=True)
```

```python
df['runs_left'] = df['runs_target']-df['current_score']+1
df['balls_left'] = 120 - df['team_balls']
df['CRR']=df.current_score*6/(120-df.balls_left)
df['RRR']=df.runs_left*6/df.balls_left
df['wickets']=10- df.team_wicket
```

```python
df.head(5)
```

```python
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
def is_win(f):
    winner = []
    for item,row in f.iterrows():
        if row.winner == row.batting_team:
            winner.append(1)
        else:
            winner.append(0)

    return winner
```

```python
df['winner']=is_win(df)
```

```python
df.head(5)
```

```python
final_df =df[['match_id','batting_team','bowling_team','city','runs_left','balls_left','wickets','runs_target','CRR','RRR','winner']]
```

```python
final_df.head(5)
```

```python
final_df.duplicated().sum()
```

```python
final_df =final_df.sample(final_df.shape[0])
```

```python
team_name_map = {
    'Gujarat Lions': 'Gujarat Titans',
    'Kings XI Punjab': 'Punjab Kings',
    'Deccan Chargers': 'Sunrisers Hyderabad',
    'Royal Challengers Bangalore': 'Royal Challengers Bangalore',
    'Pune Warriors': 'Rising Pune Supergiants',
    'Delhi Daredevils': 'Delhi Capitals'
}

final_df['batting_team'] = final_df['batting_team'].replace(team_name_map)
final_df['bowling_team'] = final_df['bowling_team'].replace(team_name_map)
```

```python
final_df.batting_team.unique()
```

```python
teams_to_remove = [
    'Rising Pune Supergiant',
    'Rising Pune Supergiants',
    'Kochi Tuskers Kerala'
]


final_df = final_df[
    ~final_df['batting_team'].isin(teams_to_remove) &
    ~final_df['bowling_team'].isin(teams_to_remove)
]
```

```python
final_df.bowling_team.unique()
```

```python
final_df.city.unique()
```

```python
final_df.dropna(inplace=True)
final_df =final_df[final_df.balls_left !=0]
```

```python
final_df.RRR.replace([np.inf,-np.inf],np.nan,inplace=True)
```

```python
final_df.CRR.replace([np.inf,-np.inf],np.nan,inplace=True)
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.info()
```

```python
final_df.duplicated().sum()
```

```python
from sklearn.model_selection import train_test_split
X = final_df.drop(columns=['winner','match_id'])
y = final_df['winner']
X_check =final_df.drop(columns=['winner'])
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
X.sample(5)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

categorical_cols = ['batting_team', 'bowling_team', 'city']

trf = ColumnTransformer(
    transformers=[
        ('cat',
         OneHotEncoder(
             drop='first',
             sparse_output=False,
             handle_unknown='ignore'
         ),
         categorical_cols)
    ],
    remainder='passthrough'
)
```

```python
pipe = Pipeline([
    ('step1', trf),
    ('step2', LogisticRegression(solver='liblinear'))
])
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"RMSE: {rmse}")
```

```python
accuracy_score(y_test,y_pred)
```

```python
prob = pipe.predict_proba(X_test)[200][1] * 100
print(f"Winning probability: {prob:.2f}%")
```

```python

from sklearn.metrics import roc_auc_score

y_prob = pipe.predict_proba(X_test)[:,1]
roc_auc_score(y_test, y_prob)

```
