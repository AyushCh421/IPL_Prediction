```python
## Filters rows where innings = 2 (team chasing)
```

```python
df_innings2 = df[df['innings'] == 2][['match_id', 'batting_team', 'runs_target']].drop_duplicates()
```

```python
## Filters rows where innings = 1 (team batting first)
```

```python
inning1_scores = df[df['innings'] == 1].groupby('match_id').agg({'team_runs': 'max'}).reset_index().rename(columns={'team_runs': 'inning1_score'})
df.head()
```

```python
## Merging both datasets
```

```python
df_new = pd.merge(df_innings2, inning1_scores, on='match_id')
```

```python
df_new = df_new.rename(columns={'runs_target': 'target_runs'})
df_new.sample(5)
```

```python
df_new.drop(columns=['target_runs','batting_team'],inplace=True)
```

```python
df=df[df['innings']==1]
df.drop(columns=['ball_no'],inplace=True)
```

```python
df=df.merge(df_new,on='match_id')
```

```python
df.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
df.head(5)
```

```python

df.drop(columns=['innings','runs_bowler','match_won_by',"runs_target"],inplace=True)
```

```python
df.rename(columns={'team_runs':'current_score'},inplace=True)
df.rename(columns={'inning1_score':'Final_score'},inplace=True)
df.rename(columns={'team_wicket':'Wickets_fallen'},inplace=True)
df.rename(columns={'overs':'Overs_left'},inplace=True)
```

```python
df["balls_left"] = 120 - df['team_balls']
df['CRR']=df.current_score*6/(120-df.balls_left)
df['wickets']=10- df.Wickets_fallen
```

```python
df.head()
```

```python
df.shape
```

```python
final_df =df[['match_id','batting_team','bowling_team','city','venue','toss_winner','toss_decision','Overs_left','balls_left','current_score','wickets','CRR','Final_score']]
```

```python
final_df.head(3)
```

```python
final_df.duplicated().sum()
```

```python
final_df = final_df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
final_df.head()
```

```python
final_df.batting_team.unique()
```

```python
team_name_map = {
    'Gujarat Lions': 'Gujarat Titans',
    'Kings XI Punjab': 'Punjab Kings',
    'Deccan Chargers': 'Sunrisers Hyderabad',
    'Royal Challengers Bangalore': 'Royal Challengers Bangalore',
    'Pune Warriors': 'Rising Pune Supergiants',
    'Delhi Daredevils': 'Delhi Capitals'
}

final_df['batting_team'] = final_df['batting_team'].replace(team_name_map)
final_df['bowling_team'] = final_df['bowling_team'].replace(team_name_map)
final_df['toss_winner']  = final_df['toss_winner'].replace(team_name_map)
```

```python
teams_to_remove = [
    'Rising Pune Supergiant',
    'Rising Pune Supergiants',
    'Kochi Tuskers Kerala'
]


final_df = final_df[
    ~final_df['batting_team'].isin(teams_to_remove) &
    ~final_df['bowling_team'].isin(teams_to_remove) &
    ~final_df['toss_winner'].isin(teams_to_remove)
]
```

```python
final_df.batting_team.unique()
```

```python
final_df.info()
```

```python
final_df.dropna(inplace=True)
final_df =final_df[final_df.balls_left !=0]
```

```python
##Checking infinity 
```

```python
final_df['CRR'] = final_df['CRR'].replace([np.inf, -np.inf], np.nan)
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.drop(columns=['venue','city'],inplace=True)
final_df.head()
```

```python
final_df.duplicated().sum()
```

```python
final_df['toss_decision'] = final_df['toss_decision'].replace({'field': 'ball'})
final_df.toss_decision.unique()
```

```python
from sklearn.model_selection import train_test_split

X = final_df.drop(columns=['Final_score', 'match_id'])
y = final_df['Final_score']

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    random_state=1,
    stratify=y ##To keep class balance.....
)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from xgboost import XGBRegressor
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestRegressor
```

```python
X.sample(3)
```

```python
categorical_cols = [
    'batting_team',
    'bowling_team',
    'toss_winner',
    'toss_decision'
]

trf = ColumnTransformer(
    transformers=[
        ('cat',
         OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore'),
         categorical_cols)
    ],
    remainder='passthrough'
)
```

```python
from xgboost import XGBRegressor

pipe = Pipeline([
    ('step1', trf),
    ('step2', XGBRegressor(
        n_estimators=500,
        learning_rate=0.1,
        max_depth=6,
        random_state=1
    ))
])
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test,y_pred))
print(f"RMSE Value: {rmse}")
```

```python
from sklearn.model_selection import cross_val_score

cv_rmse = -cross_val_score(
    pipe,
    X, y,
    cv=5,
    scoring='neg_root_mean_squared_error'
).mean()

print("CV RMSE:", cv_rmse)
```
